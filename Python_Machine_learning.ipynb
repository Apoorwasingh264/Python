```python
# =========================================
# MACHINE LEARNING PREPROCESSING
# Dataset: Employee_productivity_Dataset
# =========================================

# Import Libraries
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load Dataset
df = pd.read_csv("/content/employee_productivity_dataset.csv")

# Display first 5 rows
print("Original Dataset:\n")
print(df.head())
```

```python
## Q1. Handle Missing Values (Basic Cleaning)
## Fill missing values using the following:
## Age → Median
## Salary → Mean
## Hours_Worked_Per_Week → Median
## Performance_Score → Mean

df['Age'] = df['Age'].fillna(df['Age'].median())

df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(
    df['Hours_Worked_Per_Week'].median()
)

df['Performance_Score'] = df['Performance_Score'].fillna(
    df['Performance_Score'].mean()
)

# Display dataset after handling missing values
print("\nDataset after handling missing values:\n")
print(df.head())
```

```python
## Q2. Label Encoding
## Show the updated columns.
le = LabelEncoder()

## Convert these columns into numeric values using Label Encoding: Gender, Department
df['Gender'] = le.fit_transform(df['Gender'])

df['Department'] = le.fit_transform(df['Department'])

# Create LabelEncoder object
print("\nDataset after Label Encoding:\n")
print(df[['Gender', 'Department']].head())
```

```python
## Q3. One-Hot Encoding

## Apply One-Hot Encoding on: Work_Mode, Location
df_encoded = pd.get_dummies(df, columns=['Work_Mode', 'Location'])

## Display the dataset
print("\nDataset after One-Hot Encoding:\n")
print(df_encoded.head())

## check how many new columns are created.
print("\nTotal Columns After Encoding:", df_encoded.shape[1])
```

```python
## Q4. Normalization (Min-Max Scaling)

# Create scaler object
minmax = MinMaxScaler()

## Normalize the following columns: Salary,Hours_Worked_Per_Week
df[['Salary', 'Hours_Worked_Per_Week']] = minmax.fit_transform(
    df[['Salary', 'Hours_Worked_Per_Week']]
)

## Use MinMaxScaler and display results.
print("\nDataset after Min-Max Scaling:\n")
print(df[['Salary', 'Hours_Worked_Per_Week']].head())
```

```python
## Q5. Standardization (Scaling)

# Create StandardScaler object
scaler = StandardScaler()

## Apply StandardScaler on: Age, Projects_Completed
df[['Age', 'Projects_Completed']] = scaler.fit_transform(
    df[['Age', 'Projects_Completed']]
)

# Display transformed values
print("\nDataset after Standardization:\n")
print(df[['Age', 'Projects_Completed']].head())
```

```python
## Q6. Compare Scaling Methods

# Create copies
salary_data = df[['Salary']]

# MinMax Scaling
minmax_scaled = MinMaxScaler().fit_transform(salary_data)

# Standard Scaling
standard_scaled = StandardScaler().fit_transform(salary_data)

# Create comparison dataframe
comparison = pd.DataFrame({
    'Original_Salary': salary_data['Salary'],
    'MinMaxScaler': minmax_scaled.flatten(),
    'StandardScaler': standard_scaled.flatten()
})

# Display comparison
print("\nComparison of Scaling Methods:\n")
print(comparison.head())
```

```python
## Q7. Build Preprocessing Pipeline

# Define categorical and numerical columns

categorical_cols = ['Work_Mode', 'Location']

numerical_cols = ['Age', 'Salary', 'Hours_Worked_Per_Week',
                  'Projects_Completed', 'Performance_Score']

# ColumnTransformer
preprocessor = ColumnTransformer(

    transformers=[

        ('num', StandardScaler(), numerical_cols),

        ('cat', OneHotEncoder(), categorical_cols)

    ]
)

# Create Pipeline
pipeline = Pipeline(steps=[

    ('preprocessor', preprocessor)

])

print("\nPipeline Created Successfully")
```

```python
## Q8. Apply Pipeline

# Apply pipeline
transformed_data = pipeline.fit_transform(df)

# Display transformed dataset
print("\nTransformed Dataset:\n")
print(transformed_data)

# Shape of final dataset
print("\nShape of Final Dataset:")
print(transformed_data.shape)
```

Q9. Conceptual Question
Why is scaling important in Python?

Scaling is important because it brings all features to a similar range.
It improves machine learning model performance and helps algorithms
like KNN, Logistic Regression, and SVM work efficiently.

Q10. Conceptual Question
Why do we convert categorical data into numerical form?

We convert categorical data into numerical form because machine learning
algorithms cannot understand text values directly.
Encoding converts categories into numbers so models can process them.
