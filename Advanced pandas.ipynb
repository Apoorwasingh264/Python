```python
import pandas as pd

df = pd.read_csv("/content/Advanced_pandas_sales_dataset.csv")

df.head()
```

```python
#  Dataset Exploration - Perform the following exploratory steps:

# Display the first 10 rows
print(df.head(10))

# Display the dataset structure using info()
print(df.info())

# Find the number of rows and columns
print(df.shape)

# Identify the data types of each column
print(df.dtypes)
```

```python
# Q2 Sales Filtering Analysis

# Orders where Sales > 3500
print(df[df["Sales"] > 3500])

# Orders where Region = East and Sales > 2500
print(df[(df["Region"] == "East") & (df["Sales"] > 2500)])
```

Insights:
High sales orders contribute significantly to revenue.
East region has some high-value transactions above 2500 sales.

```python
# Q3 Customer Purchase Analysis

top_customers = df.groupby("Customer")["Sales"].sum().sort_values(ascending=False).head(5)

print(top_customers)
```

```python
# Q4 Regional Sales Performance

region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)

# Region with highest revenue
print("Highest Revenue Region:")
print(region_sales.idxmax())
```

```python
# Q5 Product Performance Analysis

product_quantity = df.groupby("Product")["Quantity"].sum()

print(product_quantity)

# Most popular product
print("Most Popular Product:")
print(product_quantity.idxmax())
```

```python
# Q6 Multi-Level Grouping

multi_group = df.groupby(["Region", "Category"])["Sales"].sum()

print(multi_group)
```

Insights:
Some regions perform better in specific categories.
Helps identify strong-performing category-region combinations.

```python
# Q7 Pivot Table Creation

pivot_table = pd.pivot_table(
    df,
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="sum"
)

print(pivot_table)

# Region selling most laptops
print("Region selling most laptops:")
print(pivot_table["Laptop"].idxmax())
```

```python
# Q8 Profitability Analysis

profit_analysis = df.groupby("Category")["Profit"].mean()

print(profit_analysis)

# Most profitable category
print("Most Profitable Category:")
print(profit_analysis.idxmax())
```

```python
# Q9 Discount Impact Study

discount_profit = df.groupby("Discount")["Profit"].mean()

print(discount_profit)
```

Insight:
Higher discounts may reduce average profit margins.

```python
# Q10 End-to-End Data Wrangling Task

# Step 1: Filter Sales > 3000
filtered_df = df[df["Sales"] > 3000]

# Step 2 & 3: Group by Region and Product and calculate total sales
grouped_data = filtered_df.groupby(["Region", "Product"])["Sales"].sum()

# Step 4: Convert to pivot table
final_pivot = grouped_data.unstack()

print(final_pivot)
```

Business Insights:
Identifies high-performing products in each region.
Helps businesses focus on profitable regional-product combinations.
Useful for sales strategy and inventory planning.
