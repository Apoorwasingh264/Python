```python
# Q1. Creating a DataFrame
# Create a Pandas DataFrame using a Python dictionary with the following columns: Product, Region, Sales, Quantity
# Add at least 5 rows of sample data and display the DataFrame.

import pandas as pd

data = {
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard"],
    "Region": ["East", "West", "North", "South", "East"],
    "Sales": [1200, 800, 1500, 700, 400],
    "Quantity": [5, 10, 7, 3, 15]}

df1 = pd.DataFrame(data)

print(df1)
```

```python
# Q2. Data Inspection

import pandas as pd
df = pd.read_csv("/Retail_dataset.csv")

# Display the first 5 rows of the dataset.
print(df.head())

# Display the last 5 rows.
print(df.tail())

# Show the structure of the dataset using info()
print(df.info())

# Display the shape of the dataset.
print(df.shape)
```

```python
# Q3. Column Selection- Select and display only the following columns: Product, Region, Sales

print(df[["Product", "Region", "Sales"]])
```

```python
# Q4. Indexing using .loc
#Using .loc, perform the following:
# Display the row with index 10.

print(df.loc[10])

# Display the Sales value of the row with index 20.

print(df.loc[20, "Sales"])
```

```python
# Q5. Indexing using .iloc- Using .iloc, display:
# The first 6 rows and first 4 columns.

print(df.iloc[:6, :4])

# The Sales value from the 3rd row.

print(df.iloc[2]["Sales"])
```

```python
# Q6. Row Slicing - Display:
# The first 10 rows of the dataset.

print(df[:10])

# Rows 15 to 25.

print(df[15:26])
```

```python
# Q.7 Filtering Data- Filter the dataset to display:

# Rows where Region = "East".

print(df[df["Region"] == "East"])

# Rows where Sales > 1500

print(df[df["Sales"] > 1500])
```

```python
# Q8. Multiple Condition Filtering- Display rows where:
# Region is West and Sales is greater than 1200

print(df[(df["Region"] == "West") & (df["Sales"] > 1200)])

```

```python
# Q9. Descriptive Statistics
# Generate summary statistics using: describe()

stats = df.describe()
print(stats)

# mean sales value

print("Mean sales:", df["Sales"].mean())

#  Maximum quantity sold

print("Maximum quantity sold:", df["Quantity"].max())

# Minimum sales value

print("Minimum sales value:", df["Sales"].min())
```

```python
# Q10. Aggregation Analysis - Using the Sales column, calculate:
# Total sales
print("Total Sales:", df["Sales"].sum())

# Average sales
print("Average Sales:", df["Sales"].mean())

# Maximum sales
print("Maximum Sales:", df["Sales"].max())

# Minimum sales
print("Minimum Sales:", df["Sales"].min())
```
